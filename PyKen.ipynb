```python
import scikits.statsmodels.api as sm
def FS_statsmodels_logreg(X,y):
		n_iter=5
		hops=20
		array=np.arange(5,100,hops)
		roc_mat=np.zeros((n_iter,array.shape[0]))
		X=np.array(X)
		y=np.array(y)
		#
		for j,prcnt in enumerate(array):
			fs=SelectPercentile(chi2,percentile=prcnt) # feature select using chi square
			#
			X_chi=fs.fit_transform(X,y)
			 
			cv=ShuffleSplit(X_chi.shape[0],n_iter=n_iter,test_size=0.3,train_size=0.7,random_state=0)
			for i,(train,test) in enumerate(cv):
				#logreg=sm.discrete_model.Logit(y[train],X_chi[train])
				logreg=sm.GLM(y[train],X_chi[train],family = sm.families.Binomial())
				model=logreg.fit()
				y_pred=logreg.predict(X_chi[test])
				#fpr, tpr, thresholds = roc_curve(y_test,y_pred)
				roc_mat[i,j] = roc_auc_score(y[test], y_pred)
		return roc_mat,cv,array
```

# $\chi^2$ feature selection with Random Forest fitting.

Calculate  the conufsion atrix, AUC ROC from 5 fold CV sets, and features selected from $\chi^2$

```python
def FS_chi_ROC(X,y,prcnt):
	#
		X=np.array(X)
		y=np.array(y)
				#
		fs=SelectPercentile(chi2,percentile=prcnt)
		X_chi=fs.fit(X,y).transform(X)
		rf=ExtraTreesClassifier(n_estimators=500,max_features=X_chi.shape[1],random_state=0)
		cv_score=cross_val_score(rf,X_chi,y,scoring='roc_auc',cv=5)
		#
		X_train, X_test,y_train, y_test=train_test_split(X_chi,y,test_size=0.33,random_state=0) # get the training and testing sets
		y_pred=rf.fit(X_train,y_train).predict(X_test)
		cm=confusion_matrix(y_pred,y_test)
		print(cm) # show confusion matrix
		return cv_score,fs.pvalues_,fs.get_support(),cm
```

Fit Random Forest on the $\chi^2$ set of features selected and get the ROC AUC for 5 fold cross validation

```python
def Iter_FS_chi_ROC(X,y):
		mask_list=[]
		fs_pvals=[]
		hops=20
		array=np.arange(5,100,hops)
		score_mat=np.zeros((5,array.shape[0]))
		for i,prcnt in enumerate(array):
			cv_score,pvalues,mask,cm=FS_chi_ROC(X,y,prcnt)
			score_mat[:,i]=cv_score
			fs_pvals.append(pvalues)
			mask_list.append(mask)
		n_features=[m.sum() for m in mask_list]
		return score_mat,mask_list,n_features,fs_pvals,array
```

```python
score_mat,mask_list,n_features,fs_pvals,array=Iter_FS_chi_ROC(X,y)
```

```python
def make_overlay_plot(roc_mat,score_mat):
		'''Make plot with both  ovelaying models resuls.'''
		mean_logreg=np.mean(roc_mat,axis=0)
		std_logreg=np.std(roc_mat,axis=0)
	
		mean_chi2=np.mean(score_mat,axis=0)
		std_chi2=np.std(score_mat,axis=0)
		#
		fig,ax1=pl.subplots()
		ax1.set_xlabel('n_features')
		ax1.set_title('ROC AUC for $ \chi^2$ feature selection')
		ax1.set_ylabel('ROC AUC')
		ax1.set_yticks(np.arange(0,1.1,0.1))
		ax1.set_ylim(0,1)
		ax1.errorbar(n_features,mean_logreg,yerr=std_logreg,fmt='b-',elinewidth='2')
		ax1.errorbar(n_features,mean_chi2,yerr=std_chi2,fmt='r-')
		pl.rcParams['figure.figsize']= 10,7  ;pl.rcParams['axes.grid']=True
        
		pl.legend(['logreg','RF'],loc='lower right')
		pl.tight_layout()
		pl.savefig('logreg_rf_chi2.pdf')
		pl.show()
    
```

```python
def ROC_best_chi(X,y):
		X=np.array(X)
		y=np.array(y)
		#
		fs=SelectPercentile(chi2,percentile=45)
		X_chi=fs.fit(X,y).transform(X)
		# Spit the set
		X_train, X_test,y_train, y_test=train_test_split(X_chi,y,test_size=0.33,random_state=0)
		#
		rf_chi2=ExtraTreesClassifier(n_estimators=500,max_features=X_chi.shape[1],random_state=0)
		probas=rf_chi2.fit(X_train,y_train).predict_proba(X_test)
		y_pred=rf_chi2.fit(X_train,y_train).predict(X_test)
		accuracy=accuracy_score(y_test,y_pred)
		roc_auc=roc_auc_score(y_test,probas[:,1])
#								
		fpr, tpr, thresholds = roc_curve(y_test, probas[:, 1])
		return fpr,tpr,thresholds,accuracy,roc_auc
        
```

```python
fpr,tpr,thresholds,accuracy,roc_auc=ROC_best_chi(X,y)
```

```python
def make_ROC(fpr,tpr,roc_auc,title,fname):
		# Plot ROC curve
		fig,ax = pl.subplots()
		pl.plot(fpr, tpr, label='ROC curve (area = %0.2f)' %( roc_auc))
		pl.plot([0, 1], [0, 1], 'k--')
		pl.xlim([0.0, 1.0])
		pl.ylim([0.0, 1.0]); ax.set_xlabel('False Positive Rate')
		pl.title(title);pl.rcParams['figure.figsize']= 10,7  ;pl.rcParams['axes.grid']=True
		pl.legend(loc="lower right");ax.set_ylabel('True Positive Rate')
		pl.tight_layout()
		pl.savefig(fname,dpi=400)
		pl.show()
```

```python
def GetRFImportance(X,y):
		""" This method will take fit a random forest and plot a barplot of the impotance of the 
			features"""
		forest=ExtraTreesClassifier(n_estimators=100,random_state=0)
		forest.fit(X,y)
		X_new=forest.fit(X,y).transform(X)
		importance=forest.feature_importances_
		indx=np.argsort(importance)[::-1] #sort in decreasing order
		feature_std=np.std([tree.feature_importances_ for tree in forest.estimators_],axis=0)
		feature_mean=np.mean([tree.feature_importances_ for tree in forest.estimators_],axis=0)
		summary_table=zip(feature_mean[indx],feature_std[indx],X.columns[indx])
		N=X.shape[1]
		width=0.65
		#labels=[feature_dict[db_term] for db_term in list(X.columns[indx])]
		#
		fig, ax =pl.subplots()

		ax.set_ylabel('scores from RF fitting')
		ax.set_title('Features by Importance')
		ax.tick_params(axis='x',labelsize=10)
		rect=pl.bar(range(N),feature_mean[indx],width,color='r',yerr=feature_std[indx])	
		pl.xticks((np.arange(len(X.columns))+0.32),(X.columns[indx]),rotation=89)
		pl.rcParams['figure.figsize']= 10,8  ;pl.rcParams['axes.grid']=True
		pl.tight_layout()
		pl.savefig('Importance2.pdf')
		pl.show()
		
```

```python
def Iter_RF_Importance(X,y):
		""" This method will iterate through the set of Important variables and fit a model with the necessary features
		 it return cv_matrix,cv_new,cm, X_new
			cv_matrix  matrix of scores 
			cv_new  tA new transfored matrix of data with the best features from scikit learn
 			X_new= the transforemd X matrix with the new features.
		"""
		n_iter=5
		max_feature=25
		#
		# Compute the importnce 
		forest=ExtraTreesClassifier(n_estimators=100,random_state=0)
		forest.fit(X,y)
		importance=forest.feature_importances_
		indx=np.argsort(importance)[::-1] #sort in decreasing order
		indx=indx[:max_feature]
		forest_new=ExtraTreesClassifier(n_estimators=100,random_state=0)
		#
		score=np.zeros((X.shape[0],2))
		cv=ShuffleSplit(X.shape[0],n_iter=n_iter,test_size=0.3,train_size=0.7,random_state=0)# generate indices
				#
		prec_matrix=np.zeros((n_iter,max_feature))  
		for features in np.arange(0,max_feature):
			forest_new=ExtraTreesClassifier(n_estimators=500,max_features=(1+features),random_state=0)
			prec_matrix[:,features]=cross_val_score(forest_new,X.ix[:,indx[0:features+1]],y,cv=cv,scoring='roc_auc')
			#print(features)
		return prec_matrix
```
